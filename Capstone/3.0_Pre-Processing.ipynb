<font color="blue">Variables & Formulas that will be used in our dataset<font>

```python
#Real Estate Variables

interest = .065/12
property_tax = .0071
maintenance = .003
insurance = .0025
closing_costs = .04
mls_amt = df['MLS Amount']
term = 360

#Real Estate Formulas

#Principal: Balance of the loan to purchas at the MLS Amount and 25% down.
principal = (mls_amt*.75) +(mls_amt*closing_costs)

#Monthly Mortgage payment of principal and interest
mortgage = principal * (interest*((1+interest)**term))/ (((1+interest)**term)-1)

#Costs: Annual cost of the mortgate, insurance, property taxes, and property maintenance
Costs = (mortgage * 12) + (mls_amt * maintenance) + (mls_amt * insurance) + (mls_amt * property_tax)
```

```python
#Commented out so not to rerun
#Formatting columns and cells for easy referencing

df.drop('county', axis=1, inplace=True)
df.columns = [x.lower() for x in df]
df.columns = df.columns.str.replace(' ', '_')
df['county'] = df['county'].replace(' ', '_', regex=True)
df['property_type'] = df['property_type'].replace(' ', '_', regex=True)
df['county'] = df['county'].replace([x.lower() for x in df])
# df['name_county'] = df['county'].copy
```

```python
#Commented out so not to rerun
#Adding our Real Estate Variables as columns

df.insert(loc=17, column = 'principal', value = principal)
df.insert(loc=18, column = 'mortgage', value = mortgage)
df.insert(loc=19, column = 'costs', value = Costs)
df.insert(loc=20, column = 'price_per_sqft', value = 0)
df['price_per_sqft'] = (df['mls_amount']/  df['building_sqft'])
```

```python
#Commented out so not to rerun
#The following lines of code are creating a DOM, Days on Market, column. This is a useful metric for investors. At this moment in the
#beginning of 2023 DOM is an increasing metric. As it increases, price trends downwards.

df.insert(loc =20, column = 'dom', value = 0)
df['dom'] = pd.Timestamp.today().strftime('%Y-%m-%d')

df.insert(loc =20, column = 'today', value = 0)
df['today'] = pd.Timestamp.today().strftime('%Y-%m-%d')

df['mls_date'] =  pd.to_datetime(df['mls_date'])
df['dom'] =  pd.to_datetime(df['dom'])
df['today'] =  pd.to_datetime(df['today'])

df['dom'] = (df['today'] - df['mls_date'])

df['today'] = pd.Timestamp.today().strftime('2023-1-11')

df['dom'] = df['dom'].dt.days
```

```python
df['property_type'].value_counts()
```

```python
#After looking at the Value Counts of Property types, I decided to drop 2/4 values.
#I also relabeled the other two Values for Plotting purposes.

df = df.loc[~((df['property_type'] == 'Residential_(General)_(Single)') | (df['property_type'] == 'Townhouse_(Residential)'))]

df['property_type'] = df['property_type'].replace('Single_Family_Residential', 'SFR', regex=True)
df['property_type'] = df['property_type'].replace('[()]', '', regex=True)
df['property_type'] = df['property_type'].replace('Condominium_Residential', 'Condo', regex=True)
```

```python
#Commented out so not to rerun
#Setting Index to Zip column for Mapping of Rental Data

rent.set_index('Unnamed: 0', inplace = True)
rent.index.rename('zip', inplace = True)

#Inserting Rental Information by mapping to the specific County and # of Beds. Turning our rental information into 
#useful data. The next 3 cells calculate annual revenue for each unit at the obsserved minimum, average, and maximum.


def transform(x):
    code = x['zip']
    beds = x['bedrooms']
    if code in rent.index:
        return rent.loc[code, f'bed{beds}_minRent']
df.insert(loc=21, column = 'min_annual_rent', value = 12*(df.apply(transform, axis = 1)))

def transform(x):
    code = x['zip']
    beds = x['bedrooms']
    if code in rent.index:
        return rent.loc[code, f'bed{beds}_averageRent']   
df.insert(loc=22, column = 'avg_annual_rent', value = 12*(df.apply(transform, axis = 1)))

def transform(x):
    code = x['zip']
    beds = x['bedrooms']
    if code in rent.index:
        return rent.loc[code, f'bed{beds}_maxRent']
df.insert(loc=23, column = 'max_annual_rent', value = 12*(df.apply(transform, axis = 1)))
```

```python
#Commented out so not to rerun
#Calculating if the listing is profitable at our 3 levels of revenue.

df.insert(loc=24, column = 'min_cfp', value = 0)
df['min_cfp'] = (df['min_annual_rent'] - df['costs']) >0

df.insert(loc=25, column = 'avg_cfp', value = 0)
df['avg_cfp'] = (df['avg_annual_rent'] - df['costs']) >0

df.insert(loc=26, column = 'max_cfp', value = 0)
df['max_cfp'] = (df['max_annual_rent'] - df['costs']) >0
```

```python
#Commented out so not to rerun
#These 3 items calculate the annual investment return on the initial investment needed to purchase the listing at the 
#current MLS Amount.

df.insert(loc=27, column = 'min_coc', value = 0)
df['min_coc'] = (df['min_annual_rent'] - df['costs'])/((closing_costs)+(df['mls_amount']*.25))

df.insert(loc=28, column = 'avg_coc', value = 0)
df['avg_coc'] = (df['avg_annual_rent'] - df['costs'])/((closing_costs)+(df['mls_amount']*.25))

df.insert(loc=29, column = 'max_coc', value = 0)
df['max_coc'] = (df['max_annual_rent'] - df['costs'])/((closing_costs)+(df['mls_amount']*.25))
```

```python
df.insert(loc=30, column = 'target_cfp', value = 0)
df['target_cfp'] = (df['avg_coc']) > .04
```
